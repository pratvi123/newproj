```python

def f(t):
    return np.exp(-t)*np.cos(2*np.pi*t)
t1=np.arange(0,5,0.1)
t2=np.arange(0,5,0.02)
plt.figure(1)

plt.subplot(211)
plt.plot(t1,f(t1), 'bo', t2, f(t2), 'k')
plt.subplot(212)
plt.plot(t2,np.cos(2*np.pi*t2),'r--')
plt.show()
```

```python
ax=plt.subplot(111)
t=np.arange(0,5,0.01)
s=np.cos(2*np.pi*t)
line, =plt.plot(t,s,lw=10)
plt.annotate('local max', xytext=(3,1.5), xy=(2,1),arrowprops=dict(facecolor='green', shrink=0.1))
plt.ylim(-2,2)
plt.xlim(0,4)
plt.grid(True)
plt.show()
```
