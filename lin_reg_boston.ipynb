```python
data['data']
```

```python
df=pd.DataFrame(data.data)
df
```

```python
df.columns=data.feature_names
df.columns
```

```python
print(df.head(20))
```

```python
plt.scatter(df['INDUS'],df['TAX'])
plt.xlabel('INDUS')
plt.ylabel('TAX')
plt.show()
```

```python
S=df['INDUS']
P=df['TAX']
print(S)
print(P)
```

```python
np.shape(S)
np.shape(P)
```

```python
S=np.transpose([S])
P=np.transpose([P])
P
S
```

```python
lm=LinearRegression()
model=lm.fit(S,P)
E=model.coef_[0][0]
E


```
